```python
def populateResult(que):
        ret = faiss_index.as_retriever()
        rdocs = ret.get_relevant_documents(que)
        print("More Suitable question")
        result = []

        

        for doc in rdocs:
            prompt = doc.page_content.split('?')[0].replace('prompt: ', '')
            response = doc.page_content.split('?')[1].replace('response: ', '')
            result.append({"question":prompt,"answer":response})
            # print(f"Question: {prompt}")
            # print(f"Answer: {response}\n")
        return result
    
    
    
    
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        (
            "system",
            "You are a helpful assistant that combines the following question answers to generate single question and answer {question_answer}.",
        ),
        ("human", "{actual_question}"),
    ]
)
chain = prompt | llm

def AskQuestion():
    que = input("Enter Query : ")
    result = populateResult(que)
    return {
        "AI" : chain.invoke(
        {
            "question_answer": result,
            "actual_question": que,
        }
    ).content,
        "related_Doc": result
        }
print(AskQuestion())
```

```python
from sentence_transformers import SentenceTransformer
from langchain.vectorstores import FAISS

# Initialize the sentence transformer model
embedding_model = SentenceTransformer('all-MiniLM-L6-v2')

# Generate embeddings for FAQ data
documents = [doc.page_content for doc in data]  # Assuming `data` contains your FAQ documents
embeddings = embedding_model.encode(documents)

# Combine texts and embeddings into tuples
text_embeddings = list(zip(documents, embeddings))

# Create the FAISS index
faiss_index = FAISS.from_embeddings(
    text_embeddings=text_embeddings,  # Combined text and embeddings
    embedding=embedding_model.encode  # Embedding function
)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from rank_bm25 import BM25Okapi

# Ensure NLTK data is downloaded
nltk.download('punkt')

# Tokenize documents for BM25
tokenized_corpus = [word_tokenize(doc.page_content) for doc in data]

# Create the BM25 model
bm25 = BM25Okapi(tokenized_corpus)

def hybrid_search(query, faiss_index, bm25):
    # Perform FAISS search
    ret = faiss_index.as_retriever()
    # rdocs = ret.get_relevant_documents(que)
    dense_results = ret.get_relevant_documents(query)
    
    # Perform BM25 keyword search
    tokenized_query = word_tokenize(query)
    bm25_results = bm25.get_top_n(tokenized_query, data, n=10)  # Retrieves top 10 results

    # Combine results by adding BM25 results to FAISS results based on similarity scores
    combined_results = dense_results[:5] + bm25_results[:5]  # Example: taking top 5 from each
    return combined_results


# # Define a hybrid search function
# def hybrid_search(query, faiss_index, bm25):
#     # Tokenize the query
#     tokenized_query = word_tokenize(query)

#     # Perform BM25 search
#     bm25_scores = bm25.get_scores(tokenized_query)
#     print(bm25_scores)
#     # Perform FAISS search
#     faiss_results = faiss_index.similarity_search(query, k=5)

#     combined_results = merge_results(faiss_results, bm25_scores)
#     return faiss_results

#     # Combine results from both models

#     # return combined_results
```
